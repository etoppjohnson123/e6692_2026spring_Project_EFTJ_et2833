```python
def compare_models_on_image(models, image, true_mask, device, threshold=0.5):

    # Move image to device
    input_tensor = image.unsqueeze(0).to(device)

    predictions = {}

    # Get prediction from each model
    with torch.no_grad():
        for name, model in models.items():

            model.eval()

            output = model(input_tensor)

            # torchvision DeepLab returns dict
            if isinstance(output, dict):
                output = output["out"]

            pred = torch.sigmoid(output)
            pred = (pred > threshold).float()

            predictions[name] = pred.cpu().squeeze()

    # Number of plots
    n_cols = 2 + len(predictions)

    plt.figure(figsize=(4 * n_cols, 4))

    plt.subplot(1, n_cols, 1)

    img = image.cpu().permute(1, 2, 0)


    img = img * torch.tensor([0.229, 0.224, 0.225]) + \
                torch.tensor([0.485, 0.456, 0.406])

    img = img.clamp(0, 1)

    plt.imshow(img)
    plt.title("Original")
    plt.axis("off")


    plt.subplot(1, n_cols, 2)

    plt.imshow(true_mask.squeeze(), cmap="gray")
    plt.title("Ground Truth")
    plt.axis("off")

    for idx, (name, pred_mask) in enumerate(predictions.items()):

        plt.subplot(1, n_cols, idx + 3)

        plt.imshow(pred_mask, cmap="gray")
        plt.title(name)
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```
